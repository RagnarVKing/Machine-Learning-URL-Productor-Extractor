```python
# Step 1: Create the LPS array (preprocessing)
def compute_lps(pattern):
    lps = [0] * len(pattern)
    length = 0  # length of the previous longest prefix which is also a suffix
    i = 1
    while i < len(pattern):
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1
    return lps


# Step 2: Search the pattern in a given text using the LPS array
def kmp_search(text, pattern):
    lps = compute_lps(pattern)
    i = 0  # index for text
    j = 0  # index for pattern
    results = []  # store the start indices of found patterns
    while i < len(text):
        if pattern[j] == text[i]:
            i += 1
            j += 1

        if j == len(pattern):
            results.append(i - j)  # found a match
            j = lps[j - 1]
        elif i < len(text) and pattern[j] != text[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1
    return results


# Example: Searching a pattern in a text
text = "ababcabcabababd\nabc"
pattern = "abc"

result = kmp_search(text, pattern)
print("Pattern found at indices:", result)
```

```python
import Levenshtein

# Funcție pentru a găsi cea mai bună categorie pentru un cuvânt dat
def best_category(word, category_file):
    # Citim categoriile din fișier
    with open(category_file, 'r') as f:
        categories = f.readlines()

    # Eliminăm caracterele newline și spațiile inutile
    categories = [cat.strip() for cat in categories]

    # Calculăm distanța Levenshtein pentru fiecare categorie
    distances = {cat: Levenshtein.distance(word, cat) for cat in categories}

    # Găsim categoria cu cea mai mică distanță
    best_cat = min(distances, key=distances.get)

    return best_cat, distances[best_cat]

# Exemplu de utilizare
word = "iphoneX"
category_file = "categorii.txt"  # fișierul cu categorii
best_cat, distance = best_category(word, category_file)

print(f"Cea mai bună categorie pentru '{word}' este '{best_cat}' cu distanța {distance}.")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Funcție pentru a găsi cea mai bună categorie pentru un cuvânt dat
def best_category_cosine(word, category_file):
    # Citim categoriile din fișier
    with open(category_file, 'r') as f:
        categories = f.readlines()

    # Eliminăm caracterele newline și spațiile inutile
    categories = [cat.strip() for cat in categories]

    # Adăugăm cuvântul la lista de categorii pentru a compara
    all_text = [word] + categories

    # Convertim textul în vectori bag of words
    vectorizer = CountVectorizer().fit_transform(all_text)

    # Calculăm cosine similarity între cuvânt și fiecare categorie
    cosine_similarities = cosine_similarity(vectorizer[0:1], vectorizer[1:])

    # Găsim categoria cu cea mai mare similitudine cosine
    best_cat_index = np.argmax(cosine_similarities)
    best_cat = categories[best_cat_index]

    return best_cat, cosine_similarities[0, best_cat_index]

# Exemplu de utilizare
word = "iphoneX"
category_file = "categorii.txt"  # fișierul cu categorii
best_cat, similarity = best_category_cosine(word, category_file)

print(f"Cea mai bună categorie pentru '{word}' este '{best_cat}' cu similitudinea cosine {similarity}.")
```
